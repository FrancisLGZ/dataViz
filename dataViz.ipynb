## Graphe 1

#### Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".

#### Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.

#### Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
#GRAPH1

#range abscisses
x = np.arange(0,50) #source : https://stackabuse.com/how-to-set-axis-range-xlim-ylim-in-matplotlib/

#dessiner le graph
plt.plot(x, x/3, color='C3')

#renommer abscisses et ordonnees
plt.xlabel('Axe des x', fontsize=10)
plt.ylabel('Axe des y', fontsize=10)
plt.title('Tracer une droite', fontsize=15)

#Voir uniquement les figures
plt.show()
```

```python
#GRAPH2

#range abscisse
x = np.arange(1,13) #source : https://stackabuse.com/how-to-set-axis-range-xlim-ylim-in-matplotlib/

#style
style = dict(color="C3", linewidth=1, linestyle="-", marker=".", mec="C7")

#dessiner le graph
plt.plot(x, df['total_profit'], **style)

#renommer abscisses et ordonnees
plt.xlabel('Mois', fontsize=10)
plt.ylabel('Profit', fontsize=10)
plt.title('Profit total par mois', fontsize=15)

#Voir uniquement les figures
plt.grid()
plt.show()
```

## Graphe 3

#### Ensuite, nous souhaitons un graphe qui donne la courbe de profit pour chaque produit, en fonction du temps.
#### Chaque courbe aura une couleur différente et nous souhaitons une légende qui indique quelle couleur correspond à quel produit.

```python
#GRAPH3
#changer la taille du graph. source : https://pythonguides.com/matplotlib-increase-plot-size/
fig, axes = plt.subplots(figsize=(15,10))

#range abscisse
x = np.arange(1,13) #source : https://stackabuse.com/how-to-set-axis-range-xlim-ylim-in-matplotlib/

#style. source : video dans la ressource.
style = dict(linewidth=2, linestyle="--", marker=".", markersize=15, markerfacecolor="C7", mec="C7")


#dessiner le graph
plt.plot(x, df['facecream'], label='Facecream', color='C1', **style)
plt.plot(x, df['facewash'], label='Facewash', color='C2', **style)
plt.plot(x, df['toothpaste'], label='Toothpaste', color='C3', **style)
plt.plot(x, df['bathingsoap'], label='Bathingsoap', color='C4', **style)
plt.plot(x, df['shampoo'], label='Shampoo', color='C5', **style)
plt.plot(x, df['moisturizer'], label='Moisturizer', color='C6', **style)

#renommer abscisses et ordonnees
plt.xlabel('Mois', fontsize=10)
plt.ylabel('Profit', fontsize=10)
plt.title('Profit total par produit et par mois', fontsize=15)

#ajout de la legende : source : https://jakevdp.github.io/PythonDataScienceHandbook/04.06-customizing-legends.html
plt.legend(title="$\\bf{PRODUITS}$")

#Voir uniquement les figures
plt.grid()
plt.show()
```
